```python

#Introduction
'''
This analysis uses the Netflix Titles dataset:
https://www.kaggle.com/datasets/shivamb/netflix-shows

The dataset contains information about movies and TV shows on Netflix,
including title, type, country, release year, rating, and duration.

This dataset was chosen because it includes both categorical and numerical
variables, as well as missing values, making it suitable for data exploration
and data wrangling using Pandas.
'''
```

```python
# Data Exploration
```

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/MKudanova/DATA602/refs/heads/main/netflix_titles.csv")
df.head()
```

```python
# check structure (columns, types)
df.info()
```

```python
# summary statistics
df.describe()
```

```python
# check missing values
df.isnull().sum()
```

```python
# Data Wrangling
```

```python
# rename column for clarity
df = df.rename(columns={'release_year': 'year'})
```

```python
# fill missing country values
df['country'] = df['country'].fillna('Unknown')
```

```python
# create new column (movie or not)
df['is_movie'] = df['type'] == 'Movie'
```

```python
# drop unnecessary column
df = df.drop(columns=['show_id'])
```

```python
# drop first row
df = df.drop(index=0)
```

```python
# filter only movies
df[df['type'] == 'Movie']
```

```python
# convert country to uppercase
df['country'] = df['country'].str.upper()
```

```python
# check if year column is numeric
pd.to_numeric(df['year'], errors='coerce').notnull().all()
```

```python
# group by type and calculate stats
df.groupby('type')['year'].agg(['mean', 'min', 'max'])
```

```python
# group by type and rating
df.groupby(['type', 'rating'])['year'].mean()
```
